```python
import pandas as pd
from sklearn import linear_model
url = 'data.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print (correlation)
```

```python
import seaborn as sns 
sns.heatmap(correlation, annot=True)
```

```python
sns.pairplot(data)
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['purchase_after_21'], data['purchase_after_21']**2]) . T
model.fit(X, y = data['user_action'])
print('Intercept cua mo hinh :',model.intercept_)
print('Coeffient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh :',model.score(X,y = data['user_action']))
import numpy as np
range = np.arange(0, 1, 0.1)
X_range = pd.DataFrame([range, range**2]).T
y_predict = model.predict(X=X_range)
data.plot(kind='scatter', x='purchase_after_21', y='user_action', figsize=(9,9), color='black')
plt.plot(range, y_predict, color='blue')
```

```python
model = linear_model.LinearRegression()
model.fit(X = data.loc[:,['purchase_after_21', 'includes_a_book']], y = data['user_action'])
print('Intercept là: ', model.intercept_)
print('Coefficent là: ', model.coef_)
print('Độ chính xác là: ', model.score(X = data.loc[:,['purchase_after_21', 'includes_a_book']],y = data['user_action']))
```
